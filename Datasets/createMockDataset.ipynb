```python
import pandas as pd
from tqdm import tqdm
import random
```

```python
def finalTouch(data, name):
    
    data = pd.DataFrame(data)
    data.columns = [name]
    data = data.sample(frac=1)
    data.reset_index(drop=True, inplace=True)
    
    return data
```

```python
'''
Dataset Size: 60000 rows
'''

numRows = 60000
numClients = 6
# Location mapping to scores. Higher means more forced labour
locations = {'Asia-Pacific': 51.8,
             'Developed Economies and EU':46.9,
             'Central and South-Eastern Europe and CIS':18.0,
             'Africa':13.1,
             'Latin America and the Caribbean':12.0,
             'Middle East':8.5
            }
```

```python
location_names = []

for i in range((int)(numRows/len(locations))):
    
    for k in locations:
        location_names.append(k)
        
location_names = finalTouch(location_names, 'Location')


# hygiene: On a scale of 5

hygiene = []

for i in range(numRows):
    
    hygiene.append(random.randint(1, 5))

hygiene = finalTouch(hygiene, 'Hygiene')

# Micromanagement: On a scale of 10

micromanage = []

for i in range(numRows):
    micromanage.append(random.randint(1, 10))
    
micromanage = finalTouch(micromanage, 'Micromanagement')

# Working Hours: On a scale of 24

workhrs = []

for i in range(numRows):
    workhrs.append(random.randint(1, 24))

workhrs = finalTouch(workhrs, 'Working Hours')

# Salary: On a scale of 100:50000

salary = []

for i in range(numRows):
    salary.append(random.randint(100, 50000))

salary = finalTouch(salary, 'Salary')

# Age: On a scale of 10 to 70 ( Children also included )

age = []

for i in range(numRows):
    age.append(random.randint(10, 70))

age = finalTouch(age, 'Age')

# Infrastructure: On a scale of 10

infra = []

for i in range(numRows):
    infra.append(random.randint(1, 10))

infra = finalTouch(infra, 'Infrastructure')

# Food Quality: On a scale of 5

food = []

for i in range(numRows):
    food.append(random.randint(1, 5))

food = finalTouch(food, 'Food Quality')

# Incentives: 

incentives = []

for i in range((int)(numRows/2)):
    incentives.append(random.randint(100, 1000))
    incentives.append(random.randint(12, 10000))

incentives = finalTouch(incentives, 'Incentives')

# Family Size : number of members in family Size: 1-10

family = []

for i in range(numRows):
    family.append(random.randint(1, 10))

family = finalTouch(family, 'Family Size')

# Experience: 1-60 years

experience= []

for i in range(numRows):
    experience.append(random.randint(1, 60))
    
experience = finalTouch(experience, 'Experience')

# Company Share Price

share = []

for i in range((int)(numRows/10)):
    
    for j in range(10):
        share.append(random.randint(2, 1500))

share = finalTouch(share, 'Share Price')

# Attrition rate %, 1-40

attrition = []

for i in range(numRows):
    
    attrition.append(random.randint(1, 40))

attrition = finalTouch(attrition, 'Attrition')

# Number of workers in contract, 10 to 10000
    
numWorkers = []
    
for i in range(numRows):
    
    numWorkers.append(random.randint(10, 10000))
    
numWorkers = finalTouch(numWorkers, 'Contract_Workers')

'''------------------------------------------------------------'''
```

```python
print("Location_Score")


location_score = []

for i in range(numRows):
    location_score.append(locations[location_names.iloc[i, 0]])

location_score = finalTouch(location_score, 'Location_Score')



print("Internal_Audit")


internal_audit = []

hMax = hygiene.max()
microMax = micromanage.max()
workHrsMax = workhrs.max()
salMax = salary.max()
ageMax = age.max()



for i in tqdm(range(numRows)):
    a = hygiene.iloc[i,0]/hMax
    b = micromanage.iloc[i,0]/microMax
    c = workhrs.iloc[i,0]/workHrsMax
    d = salary.iloc[i,0]/salMax
    e = age.iloc[i,0]/ageMax
    
    internal_audit.append(0.2*(a[0]+b[0]+c[0]+d[0]+e[0]))
    
internal_audit = finalTouch(internal_audit, 'Internal_Audit')



print("External_Audit")


external_audit = []

infMax = infra.max()
foodMax = food.max()
incenMax = incentives.max()


for i in tqdm(range(numRows)):
    
    a = infra.iloc[i,0]/infMax
    b = food.iloc[i,0]/foodMax
    c = incentives.iloc[i ,0]/incenMax
    d = age.iloc[i ,0]/ageMax
    external_audit.append(0.2*(a[0] + b[0] + c[0] + d[0]))

external_audit = finalTouch(external_audit, 'External_Audit')



print("Fin_Score")


fin_score = []

familyMax = family.max()
expMax = experience.max()

for i in tqdm(range(numRows)):
    
    a = salary.iloc[i, 0]/salMax
    b = family.iloc[i, 0]/familyMax
    c = workhrs.iloc[i, 0]/workHrsMax
    d = age.iloc[i, 0]/ageMax
    e = experience.iloc[i, 0]/expMax
    
    fin_score.append(0.2*(a[0]+b[0]+c[0]+d[0]+e[0]))
    
fin_score = finalTouch(fin_score, 'Fin_Score')



print("Loss_Score")


loss = []

shareMax = share.max()
attrMax = attrition.max()
numWorkMax = numWorkers.max()

for i in tqdm(range(numRows)):
    
    a = share.iloc[i, 0]/shareMax
    b = attrition.iloc[i, 0]/attrMax
    c = numWorkers.iloc[i, 0]/numWorkMax
    
    loss.append(0.2*(a[0]+b[0]+c[0]))
    
loss = finalTouch(loss, 'Loss_Score')


print("Past_Score")


past_score = []

for i in tqdm(range(numRows)):
    
    a = hygiene.iloc[i,0]/hMax
    b = infra.iloc[i,0]/infMax
    c = salary.iloc[i, 0]/salMax
    d = share.iloc[i, 0]/shareMax
    e = attrition.iloc[i, 0]/attrMax
    
    past_score.append(0.2*(a[0]+b[0]+c[0]+d[0]+e[0]))
    

past_score = finalTouch(past_score, 'Past_Score')
```

```python
print("isUnderRisk")


isUnderRisk = []

for i in tqdm(range(numRows)):
    
    condition = locations[location_names.iloc[i, 0]]>40
    condition = condition and workhrs.iloc[i, 0]>8
    condition = condition and age.iloc[i, 0]<18 or age.iloc[i, 0]>50
    condition = condition and attrition.iloc[i, 0]>20
    
    if condition==True:
        isUnderRisk.append(1)
    else:
        isUnderRisk.append(0)



isUnderRisk = finalTouch(isUnderRisk, 'IsUnderRisk')

cell = isUnderRisk[isUnderRisk['IsUnderRisk']==1]

dataset = pd.concat([location_names, hygiene, micromanage, workhrs, 
                    salary, age, infra, food, incentives, family,
                    experience, share, attrition, numWorkers,
                    location_score, internal_audit*50, external_audit*50,
                    fin_score*50, loss*50, past_score*50, isUnderRisk], axis = 1)

dataset.to_csv('/content/drive/MyDrive/ForcedLabour/Datasets/Forced_Labour_Dataset.csv', index=False)
```
